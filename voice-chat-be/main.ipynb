```python
import requests
import json
import re

OLLAMA_URL = "http://localhost:11434/api/chat"  # 본인의 서버 주소로 수정하세요

data = {
    "model": "qwen3:4b",
    "system": "Give me a short introduction",
    "messages": [
        {
            "role": "user",
            "content": "두 문장 이내로 음성 대화 형식으로 대답해줘: 당뇨에 좋은 음식이 뭐야?"
        }
    ],
    "stream": False
}

response = requests.post(OLLAMA_URL, json=data, stream=False)

output_text = ""

# 응답 스트리밍 처리
for line in response.iter_lines():
    if line:
        res_json = json.loads(line.decode('utf-8'))

        # 'message' 키가 있다면 message.content에서 텍스트 추출
        message = res_json.get('message', {})
        content = message.get('content', '')
        output_text += content

# <think> 태그 제거
output_text = re.sub(r'<think>.*?</think>', '', output_text, flags=re.DOTALL)

print(output_text.strip())
```
